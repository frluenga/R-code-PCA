### Análisis de Componentes Principales <a class="anchor" id="modelo"></a>


Para la aplicación del método estadístico de componentes principales (PCA), se utilizó el dataset con el cual se generó el informe sobre la felicidad en el mundo, este se encuentra publicado en el repositorio de kaggle en el link: https://www.kaggle.com/unsdsn/world-happiness. Éste clasifica a 155 países según sus niveles de felicidad, los datos fueron obtenidos de la Encuesta Mundial de Gallup en diferentes años (2015, 2016, 2017, 2018 y 2019). Para efectos de este trabajo se unificaron estos periodos y se seleccionaron 8 variables presentes en cada línea de tiempo, las cuales describen el grado en que estos factores contribuyen a evaluar la felicidad en cada país.

```r
library("dplyr")
library("ggplot2")
library("FactoMineR")
library("factoextra")
library("corrplot")
library("cluster")
library("gclus")
```

```r
dataGeneral <- read.csv("MultivariadoDataSet2.csv")
#dataGeneral %>% arrange(desc(GDP.per.capita)) #arrange(desc(Perceptions.corruption)) #%>% filter(year ==2016)
data_1 <- select(dataGeneral,-index, -Country)
head(data_1)
```

## Correlaciones

Inicialmente se realizara el cálculo de las correlaciones entre las variables, esto con el fin de determinar si existen, su intensidad y además obtener un panorama visual del dataset que permita su entendimiento de manera rápida.

```r
data.abs.g     <- abs(cor(data_1[,]))
data.colors.g  <- dmat.color(data.abs.g)
data.ordered.g <- order.single(cor(data_1[,]))
cpairs(data_1, data.ordered.g, panel.colors=data.colors.g, gap=0.5)
```

Partiendo del grafico anterior, podemos establecer la existencia de correlaciones fuera de la diagonal principal, este aspecto nos da el punto de partida para la aplicación del PCA al conjunto de datos. Obsérvese que los cuadros generados con color Rojo, nos hacen referencia a la posibilidad de que exista una correlación fuerte, de color azul en aquellas que existe una ligera relación lineal y las de color amarillo una correlación casi nula.

## Componentes Principales

A continuación se realizará el cálculo del PCA, con el fin de determinar el porcentaje de la información representada en cada una de las dimensiones generadas a partir de la función PCA()[in FactorMineR], además de lo anterior se construirá un Scree plot que permita la visualización de los resultados obtenidos.

```r
res.pca <- PCA(data_1[,], ncp=5, graph = FALSE)
#res.pca <- PCA(data_1[,], scale=TRUE, graph = FALSE)

eig.val.g <-get_eigenvalue(res.pca)
eig.val.g
#print(res.pca)
```

```r
fviz_eig(res.pca, addlabels = TRUE, ylim = c(0,50))
```

Partiendo del resultado y grafico obtenido, podemos establecer que entre la primera, segunda y tercera componente absorben el 76.59% de la varianza, no obstante si sumamos la 4 componente obtendríamos un 84.58%, aunque por si sola representa un 7.99%, esto último condicionaría su elección.

```r
var.g <- get_pca_var(res.pca)
var.g
```

```r
corrplot(var.g$cos2, is.corr = FALSE)
```

```r
#Contribuciones
fviz_contrib(res.pca, choice="var", axes=1, top=10)
fviz_contrib(res.pca, choice="var", axes=2, top=10)
fviz_contrib(res.pca, choice="var", axes=3, top=10)
```

Teniendo en cuenta los gráficos anteriores, se puede observar las diferentes contribuciones de las variables frente a las dimensiones obtenidas, siendo la primera la que  presenta un mayor grado de captura de información, específicamente en Score, GDP.per.capita, Social.support, Healthy.life.expectancy y en menor fuersa Freedom.to.make.choices , en la segunda dimensión la captura de información está presente en las variables restantes (Generosity, Perceptions.corruption y year), con lo cual se podría seleccionar estas dos, obteniendo el 64.62% de captura en la varianza, descartando las demás.


```r
fviz_pca_var (res.pca, col.var = "cos2",
             gradient.cols = c("#00AFBB", "#E7B800", "#FC4E07"),
             repel = TRUE
             )
```


-	Partiendo del hecho, entre mayor longitud presente un vector de x variable, este se encuentra mejor representado dentro de los componentes del Biplot, por esto podemos determinar que la variable year, no está bien representado en las primeras dos dimensiones dado que su vector presenta una magnitud menor en comparación a las demás, absorbiendo menos variabilidad; en contraste si observamos la variable score, esta presenta una mayor magnitud en su vector quedando muy cerca de la circunferencia, lo cual indicando que se encuentra bien representada y absorbe una mayor cantidad de la variabilidad.


-	A parte de lo anterior, la variable Score nos permite inferir que entre mayor sea su valor, las variables Per cápita, expectativa de vida y el soporte social del país son altos, en este caso, podríamos establecer que estos países presentan un grado alto de desarrollo en aspectos como salubridad, económico y social, lo cual incide de manera positiva la percepción de sus ciudadanos en torno a su felicidad.


-	Otro aspecto que podemos determinar es que al aumentar el año la generosidad del país disminuye, esto se puede estar generando por aspectos económicos y sociales que afectaron la confianza de la población en los años estudiados.

# Clasificación por año para el conjunto de datos.
El conjunto de datos que encontramos en una serie de datasets suministrados por la organización no gubernamental *Sustainable Development Solutions Network* Se encuentran divididos en años, según la temporalidad en la cual desarrollan este estudio a nivel mundial, como es de esperar, la cantidad de observaciones es determinada por la cantidad de países en los cuales se llevó a cabo el estudio, siendo un total de 156 países analizados para el 2019.

Esta cifra de observaciones es considerada suficiente sabiendo que la ONU reconoce 193 países a nivel mundial como estados soberanos, y permite analizar a detalle el comportamiento anual de las variables, con el fin de encontrar un resultado más adaptado a la actualidad y así pues, descartar en esta parte del análisis el comportamiento presentado en años anteriores. 

Con esto se busca encontrar indicios de alguna adaptación o cambios en la cultura de ciertas regiones y países a causa de la creciente globalización y la penetración de internet en estas regiones y/o país en los últimos años. Un contraste entre el análisis anterior de los últimos años y los resultados del último año servirá para tal fin.

#### 1. Selección del Dataset
Selección y filtro de las observaciones o países del año 2019.

```r
df_1 <- read.csv("MultivariadoDataSet2.csv")
df <- select(df_1,-index)
```

```r
df_2019 <- df %>% filter(year==2019)
df_2019 <- select(df_2019,-year, -Country)
```

#### 2. Matriz de correlación.
Para el análisis de correlación de las variables del subconjunto de datos del año 2019 se realiza un diagrama de correlación matricial, la diagonal principal donde se encuentra el nombre de las variables, es también donde su correlación es igual a 1, por tanto, los valores en el extremo superior derecho y extremo inferior izquierdo representa la relación lineal que existe entre las variables del subconjunto.

Se identifica que la Generosidad es la variable con menor relación lineal entre las demás variables del conjunto, para todas las demás variables, pero esto no significa que no exista algún tipo de relación entre la generosidad y las demás variables del conjunto. 

El resultado de correlación para las demás variables, se encuentran medianamente bien relacionados de manera lineal, siendo las variables; esperanza de vida, apoyo social, score, PIB per cápita que mejor se correlacionó de manera lineal entre ellas.

Por último, se encuentran dos variables (Percepción de la corrupción y libertad) donde existe una relación lineal débil, pero que aportarían al análisis exploratorio de datos del subconjunto de datos.

```r
data.abs     <- abs(cor(df_2019))
data.colors  <- dmat.color(data.abs)
data.ordered <- order.single(cor(df_2019))
cpairs(df_2019, data.ordered, panel.colors=data.colors, gap=0.5)
```

#### 3. Análisis de Componentes Principales.
#### 3.1 PCA de la librería FactoMineR: 
Para realizar el análisis de componentes principales se utiliza la función PCA() de la librería FactoMineR, la cual presenta una serie de ventajas en el desarrollo del estudio, entre esas ventajas se encuentra la estandarización interna de las observaciones, ya que el PCA depende de la escala y unidades de las variables, la varianza encontrada en alguna de ellas puede ser significativamente sobresaliente de las demás, dificultando la representación correcta de las variables, en ese orden de ideas el conjunto de datos para las variables se representan con desviación estándar 1 y media 0.

Este procedimiento se encuentra como uno de los pasos relevantes para la construcción de modelos de aprendizaje no supervisado, a diferencia de los modelos supervisados como la regresión lineal donde la estandarización no tiene un efecto resultado.

```r
my.pca <- PCA(df_2019,graph=FALSE)
```

#### 3.2 Autovalores y Autovectores
Una vez se desarrolla el PCA se extrae información importante como la varianza explicada en cada componente, se puede encontrar que las dos primeras componentes absorben los mayores porcentajes de varianza de las variables, con un total de varianza acumulada del 74.85%, siendo un valor representativo importante para las dos primeras componentes.

Para pasos posteriores, se encontrará el desarrollo del análisis bajo la observación de dos componentes principales y la aclaración del criterio bajo el cual se realiza de esta forma.

```r
eig.val <- get_eigenvalue(my.pca)
print(eig.val)
```

En primera instancia se observa en el gráfico el comportamiento de la proporción de varianza explicada en cada una de las dimensiones, se identifica que alcanza el cien porciento (100%) de la variación de las observaciones en la dimensión número 7. Si analizamos de manera proporcional la variación esperada por cada dimensión podríamos asignar un 14.28% de participación a cada dimensión, con base a esto, podemos determinar que hasta la dimensión 2, el aporte de la varianza explicada es considerablemente alto al valor esperado de cada dimensión.

```r
fviz_eig(my.pca, addlabels = TRUE,ylimit = c(0,50))
```

#### 3.3 Representación de las variables en los componentes principales.

Gracias a esta primera representación gráfica se identifica que dos variables *Perceptions.corruption,  Freedom.to.make.choices)* no son completamente representadas en el plano de dos dimensiones, teniendo en cuenta el resultado de su Cos2. Sin embargo esta representación puede ser engañosa debido a que el cos2 obtenido para estas dos variables no es significativamente menor al valor intermedio representado, estas variables arrojan un cos2 alrededor del 0.65.

```r
fviz_pca_var(my.pca, col.var="cos2",
             axes=c(1,2),
             gradient.cols = c("#00AFBB","#E7B800","#FC4E07"),
             repel= TRUE)
```

De aquí podemos encontrar que la distancia euclidiana de la variable score es la que más se acerca a uno, siendo la mejor representada, también se puede encontrar que esta tiene una relación lineal entre el score o puntaje de felicidad del país con las variables **PIB per cápita, esperanza de vida y apoyo social** siendo coherente ya que podemos asegurar que a mejor economía, mayor esperanza de vida y apoyo social, mejorando entonces el puntaje final de felicidad del país, sin embargo, podemos encontrar otra conjetura importante en este análisis exploratorio de datos, encontramos que el crecimiento en la variable **percepción de la corrupción y libertad de tomar elecciones** supone disminución en el puntaje del PIB per cápita y esperanza de vida, sin embargo, no supone una clara disminución en el puntaje de felicidad del país.

De igual manera se puede observar que el puntaje de generosidad al aumentar no supone una clara disminución en el puntaje de felicidad del país, aunque un poco más en medida si puede influir en la disminución del puntaje del PIB per cápita del país y su esperanza de vida, pero no tan significativamente.

#### 3.4 Varianza explicada y Cos2 por dimensión para cada variable
Se puede apreciar que las 6 primeras variables están bien representadas en las 2 primeras componentes, exceptuando quizá la variable percepción de la corrupción la cual se representa en menor medida

```r
var <- get_pca_var(my.pca)
```

```r
corrplot(var$cos2, is.corr=FALSE)
```

A continuación se encuentra que las dos primeras dimensiones absorben la mayor cantidad de varianza de las primeras 6 variables, a excepción de la percepción de la corrupción, la cual alcanza progresivamente el total de varianza en la dimensión 3, al igual que la libertad de tomar decisiones la cual alcanza la mayor cantidad de varianza en la dimensión 4.

```r
corrplot(var$contrib, is.corr=FALSE)
```

#### 3.5 Otra forma de analizar la contribución de las variables en dos dimensiones.

```r
par(bty='u',cex=1.6,mfrow=c(1,2))
fviz_contrib(my.pca,choice="var",axes=1,top=10)
fviz_contrib(my.pca,choice="var",axes=2,top=10)
```
